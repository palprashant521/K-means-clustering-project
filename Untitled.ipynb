# Creating K means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop(['Private','Unnamed: 0'],axis=1))
```

```python
kmeans.cluster_centers_
```

Creating a new column called cluster 1 for Private and 0 for 
public School

```python
def converter(private):
    
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

A confusion and a classification matrix to check how well 
the K means cluster has performed

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print('\n')
print(classification_report(df['Cluster'],kmeans.labels_))
```
